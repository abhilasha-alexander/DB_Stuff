```python
%sql
DROP TABLE IF EXISTS sales_bronze.current_sales_copyinto;

CREATE TABLE sales_bronze.current_sales_copyinto
(
  Date DATE,
Customer_Age INT,
Customer_Gender STRING,
Country STRING,
Product_Category STRING,
Sub_Category STRING,
Product STRING,
Order_Quantity INT,
Profit INT,
Cost INT,
Revenue INT
);
```

```python
spark.sql(f'''
COPY INTO sales_bronze.current_sales_copyinto
   FROM '/Volumes/workspace/sales_bronze/sales/sales_data1.csv'
   FILEFORMAT = CSV
   FORMAT_OPTIONS('header' = 'true',
       'header' = 'true',
        'InferSchema' = 'true',
        'mergeSchema' = 'true',
        'dateFormat' = 'd-M-yyyy') 

        '''
).display()
```

```python
%sql
SELECT * FROM sales_bronze.current_sales_copyinto limit 10;
```

```python
%sql
DROP TABLE IF EXISTS sales_bronze.employee;

CREATE TABLE sales_bronze.employee
(
ID INT,
FirstName STRING,
Country STRING,
Role STRING
);
```
